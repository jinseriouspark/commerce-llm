```python
import pandas as pd
import numpy as np
import urllib
import requests
```

```python
with open('/content/drive/MyDrive/commerce-llm/env/cred.txt', 'rb') as f:
  cred_data = f.read().decode('utf-8').split('\n')
```

```python
client_id, client_secret = cred_data[0], cred_data[1]
```

```python
query_keyword = urllib.parse.quote('노트북케이스')
url = 'https://openapi.naver.com/v1/search/shop.json?query=' + query_keyword
```

```python
request = urllib.request.Request(url)
request.add_header('X-Naver-Client-Id', client_id)
request.add_header('X-Naver-Client-Secret', client_secret)
response = urllib.request.urlopen(request)

rescode = response.getcode()

if (rescode == 200):
  response_body = response.read()
  print(response_body.decode('utf-8'))
else:
  print('Error Code:' + rescode)
```

```python
def get_search_url(api_node, search_text, start_n, display_n):
  base = 'https://openapi.naver.com/v1/search'
  node = '/' + api_node + '.json'
  parameters = '?query=' + urllib.parse.quote(search_text) + '&start=' + str(start_n) + '&display=' + str(display_n)
  url = base + node + parameters
  return url
```

```python
get_search_url('shop', '노트북케이스', 1, 10)
```

```python
import datetime
import json

def get_result_onpage(url):
  request = urllib.request.Request(url)
  request.add_header('X-Naver-Client-Id', client_id)
  request.add_header('X-Naver-Client-Secret', client_secret)
  response = urllib.request.urlopen(request)
  print(f'{datetime.datetime.now()} Url Request Success')
  return json.loads(response.read().decode('utf-8'))
```

```python
def delete_tag(input_str):
  input_str = input_str.replace('<b>','')
  input_str = input_str.replace('</b>','')
  input_str = input_str.replace('\xa0','')
  return input_str
```

```python
rawdata = get_result_onpage(get_search_url('shop', '노트북케이스', 1, 10))
```

```python
def get_fields(json_data) :
  title = [delete_tag(d['title']) for d in json_data['items']]
  link = [d['link'] for d in json_data['items']]
  image = [d['image'] for d in json_data['items']]
  mallName = [d['mallName'] for d in json_data['items']]
  lprice = [d['lprice'] for d in json_data['items']]
  brand = [d['brand'] for d in json_data['items']]
  maker = [d['maker'] for d in json_data['items']]
  productId = [d['productId'] for d in json_data['items']]
  category1 = [d['category1'] for d in json_data['items']]
  category2 = [d['category2'] for d in json_data['items']]
  category3 = [d['category3'] for d in json_data['items']]
  productType = [d['productType'] for d in json_data['items']]

  result = pd.DataFrame({
      'title': title,
      'link': link,
      'image': image,
      'mallName': mallName,
      'lprice': lprice,
      'brand': brand,
      'maker': maker,
      'productId': productId,
      'category1': category1,
      'category2': category2,
      'category3': category3,
      'productType': productType
  }, columns = ['title','link','image','mallName', 'lprice', \
                'brand', 'maker', 'productId', 'category1', 'category2', 'category3', 'productType'])
  return result




```

```python
result_data = []
for i in range(1,10):
  rawdata = get_result_onpage(get_search_url('shop', '노트북케이스', i, 10))
  result_data.append(get_fields(rawdata))
  print(f'{datetime.datetime.now()} Page {i} Success')

result_data_concat = pd.concat(result_data)
result_data_concat.reset_index(drop=True, inplace=True)
result_data_concat
```
